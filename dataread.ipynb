```python
import random
import json
import time
import datetime
```

```python
data=[]
def data_gen():
    for _ in range(0,10):
        data.append({'timestamp':datetime.datetime.now().isoformat(),
                     'temperature':random.uniform(20,30),
                     'humidity':random.uniform(30,40)})
    return data
```

```python
interval =5
for i in range(0,10):
    with open('data.json','w') as file:
        ruth=data_gen()
        json.dump(ruth,file,indent=4)
        print(ruth)
    time.sleep(interval)    
```
